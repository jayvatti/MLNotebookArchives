```python
import pandas as pd
import numpy as np
```

```python


# Create an array of x values
x = np.linspace(-10, 10, 100)

# Calculate the corresponding y values
y = x**2 + 2*x + 3

# Create a DataFrame
df = pd.DataFrame({
    'x': x,
    'y': y
})

# Write to a CSV file
df.to_csv('../data/polyRegression.csv', index=False)
```

```python
data = pd.read_csv('../data/polyRegression.csv')
```

```python
X = np.array(data['x'])
Y = np.array(data['y'])

x_reshaped = X.reshape(-1, 1)
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# reshaping X to be a 2D array
x_reshaped = X.reshape(-1, 1)

# creating a PolynomialFeatures object with degree 2
poly = PolynomialFeatures(degree=2)

# transforming the x data [0 degree, 1 degree ,2 degree] since the degree is 2 
x_poly = poly.fit_transform(x_reshaped)

# creating a LinearRegression object
model = LinearRegression()

# training the model
model.fit(x_poly, Y)

# predictions
y_pred = model.predict(x_poly)

# plotting the original data
plt.scatter(X, Y, label='Original Data')

# plotting the polynomial curve
plt.plot(X, y_pred, color='red', label='Polynomial Curve')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Original Data with Polynomial Curve')

plt.legend()

plt.show()
```
